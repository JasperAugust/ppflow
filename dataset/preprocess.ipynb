```python
import os
bench_lig = [filename for filename in os.listdir('/gpfs/helios/home/tootsi/homing/ppflow/dataset/PPDbench/ligands133') if filename.endswith('pdb')]
bench_rec = os.listdir('/gpfs/helios/home/tootsi/homing/ppflow/dataset/PPDbench/receptors133')
import shutil
```

```python
new_path = '/gpfs/helios/home/tootsi/homing/ppflow/dataset/PPDbench/'
bench_name_list = []
for name in bench_lig:
    print(name)
    pdb_name = name.split('.')[0]
    dir_name = os.path.join(new_path, pdb_name)
    os.makedirs(dir_name, exist_ok=True)
    ligand_from = os.path.join('/gpfs/helios/home/tootsi/homing/ppflow/dataset/PPDbench/ligands133', name)
    ligand_to = os.path.join(dir_name, 'peptide.pdb')
    receptor_from = os.path.join('/gpfs/helios/home/tootsi/homing/ppflow/dataset/PPDbench/receptors133', pdb_name + '.recep.pdb')
    receptor_to = os.path.join(dir_name, 'receptor.pdb')
    shutil.copyfile(ligand_from, ligand_to)
    shutil.copyfile(receptor_from, receptor_to)
    bench_name_list.append(pdb_name)
import torch
torch.save(bench_name_list, '/gpfs/helios/home/tootsi/homing/ppflow/dataset/PPDbench/pdb_benchmarked.pt')
```
